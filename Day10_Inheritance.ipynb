```python
# MultiLevel Inheritance
class A:
    def __init__(self, name):
        self.name = name
    def get(self):
        return self.name

class B(A):
    def __init__(self, name, age, salary):
        super().__init__(name)   # Call A's constructor
        self.age = age
        self.salary = salary
    def get_info(self):
        return self.age, self.salary, self.name

class C(B):
    def __init__(self, name, age, salary, address):
        super().__init__(name, age, salary)   # Call B's constructor
        self.address = address
    def add_details(self):
        return self.name, self.age, self.salary, self.address

c = C("John", 30, 50000, "New York")
print(c.get())         
print(c.get_info())     
print(c.add_details())  
```

```python
# MultiLevel Inheritance
class Grandparent:
    def __init__(self, bhk, sqrft):
        self.bhk=bhk
        self.sqrft=sqrft
    def house(self):
        print("Garndparent House")
        
class Parent(Grandparent):
    def __init__(self,wheels,colorBrand, bhk, sqrft):
        self.wheels=wheels
        self.colorBrand=colorBrand
        super(). __init__(bhk,sqrft)
    def car(self):
        print("Parent Car")

class Child(Parent):
    def __init__(self,company,cc,mileage,wheels,colorBrand,bhk,sqrft):
        self.company=company
        self.cc=cc
        self.mileage=mileage
        super(). __init__(wheels,colorBrand,bhk,sqrft)
    def bike(self):
        print("Child Bike")
    def info(self):
        return self.company,self.cc,self.mileage,self.wheels,self.colorBrand,self.bhk,self.sqrft

bhk=input('enter number of bhk')
sqrft=input('enter number of sqrft ')
wheels=input('enter number of wheels')
colorBrand=input('enter number of colorBrand')
company=input('enter company')
cc=input('enter number of cc')
mileage=input('enter number of mileage')
c = Child(company,cc,mileage,wheels,colorBrand,bhk,sqrft)
c.house()         
c.car()     
c.bike()  
print(c.info())
```

```python
class person:
    def __init__(self,name):
        self.name=name
    def display_name(self):
        print(f" NAME is: {self.name}")
        
class student(person):
    def __init__(self,course,name):
        self.course=course
        super(). __init__(name)
    def display_course(self):
        print(f" COURSE is: {self.course}")
        
        
class collegeStudent(student):
    def __init__(self,year,course,name):
        self.year=year
        super(). __init__(course,name)
    def display_year(self):
        print(f" YEAR is: {self.year}")
        

c=collegeStudent(2005,'DataScience','Divya')
c.display_name()
c.display_course()
c.display_year()
```

```python
#Multiple Inheritance
class A:
    def func1(self):
        print('Belongs to class A')
class B():
    def func2(self):
        print('Belongs to class B')
class C(A,B):
    def func3(self):
        print('Belongs to class C')
         
c=C()
c.func1()
c.func2()
```
