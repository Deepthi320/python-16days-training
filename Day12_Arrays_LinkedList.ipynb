```python
#What is an array?
#List VS Array
#An array is collection of elements stored in continuous memory location.
#All elements are usually of same data type.
#Accesses using index(Starts from 0).
```

```python
#Index:  0  1  2  3  4
#Array:[10,20,30,40,50]

#single values are stored directly
#but list values are denoted by one variable are stored in heap memory(object reference)

#Arrays support operations like:
#Insertion, Deletion, Traversal, Searching, Updating elements
```

```python
#finding maximum number
s=[10,29,99,35,78]
max_val=s[0]
for i in s:
    if (i > max_val):
        max_val = i
print('Max_val:',max_val)
```

```python
#finding minimum number
s=[10,29,99,35,8]
min_val=s[0]
for i in s:
    if (i < min_val):
        min_val = i
print('Min_val:',min_val)
```

```python
#Reverse
arr = [10, 29, 99, 35, 78, 56]
n = len(arr)

for i in range(n // 2):
    arr[i], arr[n-i-1] = arr[n-i-1], arr[i]

print("Reversed Array:", arr)
```

```python
#insertion
def insert(arr, index, value):
    n = len(arr)
    new_arr = [0] * (n + 1)  # create a new array with +1 size

    for i in range(n + 1):
        if i < index:
            new_arr[i] = arr[i]
        elif i == index:
            new_arr[i] = value
        else:
            new_arr[i] = arr[i - 1]
    
    return new_arr

arr = [10, 20, 30, 40]
arr = insert(arr, 2, 25)  # insert 25 at index 2
print(arr) 
```

```python
#Deletion
def delete(arr, index):
    n = len(arr)
    new_arr = [0] * (n - 1)  # create a new array with +1 size

    for i in range(n):
        if i < index:
            new_arr[i] = arr[i]
        elif i > index:
            new_arr[i-1] = arr[i]
    return new_arr

arr = [10, 20, 30, 40]
arr = delete(arr, 2)  # insert 25 at index 2
print(arr)
```

```python
#Updating
def update(arr, index, value):
    arr[index] = value
    return arr

arr = [10, 20, 30, 40]
arr = update(arr, 1, 22)  # update index 1 to 22
print(arr)
```

```python
#What is Linked List?
# Alinked list in Python is a linear data structure made up of nodes, where each node contains:
#1.Data (the value)
#2.A reference (link) to the next node in the sequence
```

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

n1=Node(10)
print(n1.data)
```

```python
#Program for Insertion at Front
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

# New node to insert at front
new_node = Node(5)

new_node.next = head
head = new_node

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Program for Insertion at Last
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

# New node to insert at last
new_node = Node(50)

current = head
while current.next is not None:
    current = current.next

current.next = new_node

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Program for Insertion at Given Position
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1
# New node
new_node = Node(25)
position = 3

current = head
count = 1

# Move to (position-1) node
while count < position - 1:
    current = current.next
    count += 1

# Insert node
new_node.next = current.next
current.next = new_node

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Deletion at Front
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

# Delete first node
head = head.next

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Deletion at Last (End)
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

# Delete last node
current = head

while current.next.next is not None:
    current = current.next

current.next = None

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Deletion at Given Position
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

position = 3

current = head
count = 1

# Move to (position-1) node
while count < position - 1:
    current = current.next
    count += 1

# Delete node
current.next = current.next.next

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```

```python
#Deletion by Value
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Existing nodes
n1 = Node(10)
n2 = Node(20)
n3 = Node(30)
n4 = Node(40)

n1.next = n2
n2.next = n3
n3.next = n4

head = n1

# Value to delete
value = 30

# If head needs to be deleted
if head.data == value:
    head = head.next
else:
    current = head
    while current.next is not None and current.next.data != value:
        current = current.next

    if current.next is None:
        print("Value not found")
    else:
        current.next = current.next.next

# Display
current = head
while current is not None:
    print(current.data, end=" -> ")
    current = current.next

print("None")
```
