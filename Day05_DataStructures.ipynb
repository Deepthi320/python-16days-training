```python
##comprehensions
n=int(input("enter a number:"))
for i in range(1,n):
    print(i**2,end=' ')
```

```python
# syntax for comprehensions
#expression iterable if condition needed
n=int(input("enter a number:"))
l=[i**3 for i in range(1,n) if i**3%2==0]
l
```

```python
#string using comprehensions
l=['apple','banana','muskmelon']
result=[ch for ch in l]
result
```

```python
#length ofstring using comprehensions
l=['apple','banana','muskmelon']
result=[len(ch) for ch in l]
result
```

```python
#convert string to uppercase using comprehensions
l=['apple','banana','muskmelon']
result=[ch.upper() for ch in l]
result
```

```python
#Even or Odd dictionary
nums=[1,2,3,4,5]

result={x: "even"  if x%2==0 else "odd" for x in nums}

print(result)
```

```python
#Name and length dictionary
names=['Rahul','AnwarSir','Pavani']
result=[len(ch) for ch in names]
result
```
