```python
#palindrome without using slicing
text = input("Enter a string or number: ")

reverse = ""

for i in text:
    reverse = i + reverse

if text == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#palindrome using slicing
text = input("Enter a string or number: ")

if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Armstrong
num=int(input("enter a number"))
temp=num
digits=len(str(num))
total=0
for i in str(num):
    total=total+int(i)**digits
if total==temp:
    print("Armstrong")
else:
    print("Not armstrong")
```

```python
# EXAMPLE STRING
s = 'aGHInnjtht  yuyr8yH u@#$%0309*'

# initial characters
letters = 0
digits = 0
special_char = 0

# loop
for c in s:
    ascii_val = ord(c)

    if (65 <= ascii_val <= 90) or (97 <= ascii_val <= 122):
        letters += 1
    elif (48 <= ascii_val <= 57):
        digits += 1
    else:
        special_char += 1

print("Letters:", letters)
print("Digits:", digits)
print("Special characters:", special_char)
```

```python
#sum of matrix elements
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
total=0
for row in matrix:
   for elements in row:
       total=total+elements
       print(elements, end=' ')
print("total=",total)
```

```python
#Even numbers in matrix
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
total=0
for row in matrix:
   for elements in row:
       if(elements%2==0):
           total+=1
           
print("even=",total)  
```

```python
#product of matrix elements
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
total=1
for row in matrix:
   for element in row:
       total=total*element
      
print("total=",total)
```

```python
#sum of digits
n=int(input("enter a number"))
temp=abs(n)
total=0

while temp>0:
    digit=temp%10
    total+=digit
    temp=temp//10

print(f"sum of digits:{total}")
```

```python
#prime numbers
num = int(input("Enter a number: "))

i = 2
is_prime = True

if num > 1:
    while i < num:
        if num % i == 0:
            is_prime = False
            break
        i += 1

    if is_prime:
        print(num, "is a Prime Number")
    else:
        print(num, "is Not a Prime Number")
else:
    print(num, "is Not a Prime Number")
```
