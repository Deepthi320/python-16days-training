```python
#QUESTION:

#Design a Car class where each car has brand and model as instance variables:
#Create a method display_details() to print full car information.
#Create at least two car objects and display their details.
#Explain the difference between object data and class structure.

#1)CAR SHOWROOM MANAGEMENT

class Car:
    
    # Constructor
    def __init__(self, brand, model): 
        self.brand = brand      # Instance variable
        self.model = model      # Instance variable
    
    # Method to display details
    def display_details(self):
        print("Car Brand :", self.brand)
        print("Car Model :", self.model)
        print("---------------------")

# Creating objects
car1 = Car("Toyota", "Innova")
car2 = Car("Hyundai", "Creta")
# Display details
car1.display_details()
car2.display_details()
```

```python
#TEACHER SUBJECT ALLOCATION
class Teacher:
    def __init__(self, name, subject):
        self.name = name
        self.subject = subject

    def display(self):
        print("Teacher Name:", self.name)
        print("Allocated Subject:", self.subject)
        print("-------------------------")

# Creating teacher objects
t1 = Teacher("Anita", "Mathematics")
t2 = Teacher("Ravi", "Physics")
t3 = Teacher("Meena", "Computer Science")
# Display details
t1.display()
t2.display()
t3.display()
```

```python
#Static methods
#No self,No cls,defined using @static method,
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        
    def update_marks(self, new_marks):
        self.marks = new_marks   # updating marks

    def display_details(self):
        print(f"Student name: {self.name}, Current Marks: {self.marks}")

a1 = Student('Deepthi', 98)
a1.update_marks(100)

a2 = Student('Vishruth', 97)
print(a1.name, a1.marks)
print(a2.name, a2.marks)
```

```python
#class method:
class Bank:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)
        
    def withdraw(self, amount):
        self.balance -= amount
        print("Withdrawn:", amount)

c1 = Bank('Deepthi', 40000)
c2 = Bank('Rahul', 78000)
c1.deposit(5000)
c1.withdraw(4000)
print("Deepthi Balance:", c1.balance)
print("Rahul Balance:", c2.balance)
```

```python
class Car:
    def __init__(self, brand, mileage):
        self.brand = brand
        self.mileage = mileage

    def drive(self, km):
        self.mileage += km

    def show_details(self):
        print(f"Car Brand: {self.brand}, Mileage: {self.mileage}")

c1 = Car('BMW', 90)
c2 = Car('Ferrari', 85)
c1.drive(789)

c1.show_details()
c2.show_details()
```
