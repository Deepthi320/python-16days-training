```python
class A:
    def __init__(self,name):
        self._name = name   # Protected variable

a = A("DEEPTHI")
print(a._name)   # Accessing variable (no brackets)
#We can access it by 2 ways
#1> print protected variable with_ as well
#use getters to call
```

```python
#Q1
class Employee:
    def __init__(self,salary):
        self._salary=salary
    def display_salary(self):
        print("protected salary:",self._salary)

a=Employee(50000)
a.display_salary()
```

```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        self._balance -= amount
        print("Withdrawn:", amount)

    def show_balance(self):
        return self._balance


b = BankAccount(34000)
b.deposit(9000)
b.withdraw(8000)
print("Current Balance:", b.show_balance())
```
