```python
#An exception in a run time error interrupts the normal flow of a program.
#Example:
#Dividind by zero
#Accesing invalid index
#Converting strint to int

#If we don't handle eceptions->Program crashes
#If we handle them properly->Program continues safely.
```

```python
#Why exception handling is important?
#Prvents program crash
#Improves robustness(strong foundation)
#Helps in debugging
#Maintains normal program flow
```

```python
#Exception handling blocks in python
#main blocks:
#try->code that may cause error is written here.
#except->Handles the error.
#else->try & else are related, if tryBlock excutes elseBlock also gets excutes(exceptBlock excutes then tryBlock & elseBlock won't excute)
#finally->always excute whether the error occurs or not.
```

```python
#Basic synatx
#try:
   #risky code
#except someError:
   #handling code
#else:
   #runs if no error
#finally
   #always excute whether the error occurs or not
```

```python
try:
    age=int(input('Enter your age:'))
    print(age)
except ValueError:
    print('Enter numerical value')
```

```python
try:
   num1=int(input('Enter any num:'))
   num2=int(input('Enter any num:'))
   result=num1/num2
   print(result)
except ZeroDivisionError:
   print("Enter Values Greater Than Zero")
```

```python
try:
   num1=int(input('Enter any num:'))
   num2=int(input('Enter any num:'))
   result=num1/num2
   print("result:",result)
except ZeroDivisionError:
   print("Enter Values Greater Than Zero")
except ValueError:
    print('Enter numerical value')
else:
    new=result+12000
    print("New:",new)
```

```python
try:
   num1=int(input('Enter any num:'))
   num2=int(input('Enter any num:'))
   result=num1/num2
   print("result:",result)
except ZeroDivisionError:
   print("Enter Values Greater Than Zero")
except ValueError:
    print('Enter numerical value')
else:
    new=result+12000
    print("New:",new)
finally:#excutes always
    print("Execution is completed")
```

```python
try:
    file=open("sample.txt","r")
    content=file.read()
    print(content)
except FileNotFoundError:
    print("Enter Valid File Name")
finally:
    print("Execution Completed")
```

```python
#Exception error->It is used whe we don't know the particular error
try:
    num=int(input("enter a number:"))
    result=100/num
    print("Result:",result)
except Exception:
    print("Something Error")
```

```python
try:
    d={
        'name':'abc',
        'age':56
      }
    print(d['brand'])
except Exception:
    print('Error')
```

```python
#Index error
try:
   lst=[10,29,38]
   index=int(input("Enter index:"))
   print(lst[index])
except Exception:
    print('ArrayIndexOutOfBoundException')
```

```python
#Key Error
try:
   data={"name":"deepthi","age":20}
   key=input("enter element:")
   print(data[key])
except Exception:
    print('KeyNotFound')
```

```python
#Type Error(adding int and string)
try:
    a = 10
    b = "5"
    print(a + b)
except TypeError:
    print("TypeError occurred")
```

```python
#Attribute Error
try:
    num = 100
    num.append(10)
except AttributeError:
    print("AttributeError occurred")
```

```python
def calc():
   try:
       a=int(input("enter a number:"))
       b=int(input("enter a number:"))
       result=a/b
       print(result)
   except ZeroDivisionError:
       print("ZeroDivisionError")
   finally:
       print("Program is executed")
calc()
```
