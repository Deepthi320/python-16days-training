```python
#*kwargs
def shopping(**orders):
    total=0
    for product,prize in orders.items():
        total+=prize
        print(f"{product}={prize}")
    print("Total=",total)
shopping(perfume=400,soap=477, eyeliner=900, conditioner=1000)
```

```python
#factorial
def factorial(num):
    if num < 0:
         return"factorial not defined for negative numbers"

    result=1
    i=1

    while i<=num:
          result*=i
          i+=1
         
    return result

factorial(5)
```

```python
#sum of digits using functions
def sum_digits(num):
    total=0
    temp=abs(num)
    while(temp>0):
        total+=temp%10
        temp//=10
    return total

num=int(input("enter a number:"))
print(f"sum of digits:{sum_digits(num)}")
```

```python
#PRIME NUMBERS
def is_prime(num):
    if(num<=1):
        return False

    for i in range(2,num):
        if num%i==0:
            return False
    return True   
num=int(input("enter a number:"))

if is_prime(num):
    print("prime number")
else:
    print("not a prime number")
```

```python
#ARMSTRONG
def is_armstrong(num):
    temp = abs(num)
    digits = len(str(temp))
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == abs(num)


num = int(input("Enter a number: "))

if is_armstrong(num):
    print("Armstrong")
else:
    print("Not Armstrong")
```
