```python
#Nested if's
num=int(input("enter your num:"))
if(num>0):
    print("The number is positive")
    if(num%2==0):
        print("Even")
    else:
        print("Odd")
else:
    print("The number is negative")
    if(num==0):
        print("Num is zero")
    else:
        print("Num is Nonzero")
```
