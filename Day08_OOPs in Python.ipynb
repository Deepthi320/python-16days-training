```python
class Company:

    def __init__(self, name, id, salary):
        self.name = name
        self.id = id
        self.salary = salary

    def get_info(self):
        return self.name, self.id, self.salary

    def add_salary_hike(self, data):
        if self.salary >= 50000:
            hike = self.salary + data
            print("Old Salary:", self.salary)
            print("New Salary:", hike)

        elif self.salary >= 25000:
            hike = self.salary + data
            print("Old Salary:", self.salary)
            print("New Salary:", hike)

        else:
            print("No hike applicable")

# Object creation
c1 = Company("Deepthi", 40, 50000)
# Method call
c1.add_salary_hike(6000)
```

```python
#CAR SHOWROOM MANAGEMENT
#QUESTION:

#Design a Car class where each car has brand and model as instance variables:
#Create a method display_details() to print full car information.
#Create at least two car objects and display their details.
#Explain the difference between object data and clas structure.

#1)CAR SHOWROOM MANAGEMENT

class Car:
    
    # Constructor
    def __init__(self, brand, model): 
        self.brand = brand      # Instance variable
        self.model = model      # Instance variable
    
    # Method to display details
    def display_details(self):
        print("Car Brand :", self.brand)
        print("Car Model :", self.model)
        print("---------------------")

# Creating objects
car1 = Car("Toyota", "Innova")
car2 = Car("Hyundai", "Creta")
# Display details
car1.display_details()
car2.display_details()
```
